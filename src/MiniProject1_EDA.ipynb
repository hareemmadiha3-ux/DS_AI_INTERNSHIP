## Handling Missing Values

I first examined the dataset to identify columns containing missing values and calculated their percentage impact.

Columns with a very high percentage of missing data (greater than 40–50%) were considered unreliable and were dropped, as imputing a large portion could introduce bias.

For numerical columns with small percentages of missing data, I replaced missing values with the median. The median is preferred over the mean because it is less sensitive to extreme values.

For categorical columns, missing values were replaced with the mode (most frequent value) to maintain consistency in category distribution.

This approach preserves data integrity while minimizing unnecessary data loss.

## Duplicate Removal

Duplicate rows were identified to prevent repeated customer records from skewing analysis results.

Any duplicate entries were removed to ensure data accuracy and reliability.

Maintaining unique customer records ensures that statistical summaries and visualizations reflect true patterns in the dataset.

```python
plt.figure()
plt.hist(df['Age'], bins=10)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

The age distribution shows that most customers fall within the middle-age group.
There are fewer very young and very old customers, indicating the business primarily attracts adults.

```python
plt.figure()
plt.hist(df['AnnualIncome'], bins=10)
plt.title("Distribution of Annual Income")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.show()
```

The annual income distribution appears moderately spread out.
Most customers fall within a mid-income range, with fewer customers in extremely high-income brackets.

```python
plt.figure()
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

The gender distribution shows whether the customer base is balanced or skewed.
If one gender dominates, it may indicate a targeted or niche market segment.

```python
plt.figure()
plt.scatter(df['AnnualIncome'], df['SpendingScore'])
plt.title("Annual Income vs Spending Score")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

The scatter plot shows how spending behavior relates to income.
If points are widely scattered, it suggests income alone does not determine spending habits.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Gender', y='SpendingScore', data=df)
plt.title('SpendingScore by Gender')
plt.show()
```

The boxplot compares spending behavior between genders.
Differences in median or spread may indicate varying purchasing tendencies across gender groups.

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
correlation_matrix = numeric_df.corr()
```

```python
correlation_matrix
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

## Multivariate Analysis

The correlation heatmap helps identify relationships between multiple numerical variables simultaneously.

Strong positive correlations (values close to +1) indicate that two variables increase together.
Strong negative correlations (values close to -1) indicate an inverse relationship.

Weak correlations (values near 0) suggest little to no linear relationship between variables.
